# ***Q.4) Most poplar car colors***

```python
data_color = df['color'].value_counts()
print(data_color)
data_color.plot(x = 'color', y = 'value', kind = 'bar', figsize = (15,8))
```

# ***Q.5) Which car brand having maximum average price***

```python
brand_price = df.groupby('brand').mean()['price'].sort_values(ascending = False)
print(brand_price)
brand_price.plot.bar(figsize = (15,8))
```

# ***Q.6) Which state is having most expensive cars***

```python
plt.figure(figsize = (20,10))
df.groupby('state').mean()['price'].sort_values(ascending = False).plot.bar()
plt.show()
```
